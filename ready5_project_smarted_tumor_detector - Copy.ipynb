# Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer
d=load_breast_cancer()
```

# Create DataFrame from data and add target column

```python
dataset=pd.DataFrame(data=d.data, columns=d.feature_names)
dataset['target']=d.target
```

#  Data Preparation

# Separate features and target

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

# Train-test split (70% train, 30% test)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

#Prediction

```python
y_test_pred=lr.predict(x_test)
```

```python
y_test_pred
```

```python
y_test
```

# Compare actual vs predicted (XOR mismatches)

```python
print(y_test^y_test_pred)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
ac_score=accuracy_score(y_test,y_test_pred)
print(f"Accuracy score= {ac_score}")
conf_matrix=confusion_matrix(y_test,y_test_pred)
print(f"confusion matrix= {conf_matrix}")
class_report=classification_report(y_test,y_test_pred)
print(f"classification report= {class_report}")
```
